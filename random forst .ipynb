```python
import pandas as pd
import numpy as np
df = pd.read_csv("hf://datasets/phihung/titanic/train.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
```

```python
df.isnull().sum()
```

```python
df['Deck'] = df['Cabin'].fillna('Unknown').str[0]
df['Embarked'] = df['Embarked'].fillna('Unknown').str[0]
```

```python
df.isnull().sum()
```

```python
print(df["Deck"])
```

```python
df.head()
```

```python
df=df.drop(columns=["Cabin","Name"])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Sex"]=le.fit_transform(df["Sex"])
```

```python
df
```

```python
# from sklearn.preprocessing import onehotencorder
```

```python
df = pd.get_dummies(df, columns=['Ticket', 'Deck', 'Embarked'], prefix=['Ticket', 'Deck', 'Emb'],dtype=int)
```

```python
df
```

making the model to predict the data

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
x=df.drop(columns=["Survived"])
y=df["Survived"]
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape
y_test.shape
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

checking the accuracy of the model

```python
model.score(x_test,y_test)
```

checking the accuracy throught the accuracy fuinction

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
accuracy_score(y_test,y_pred)
```

making the report of the data

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
